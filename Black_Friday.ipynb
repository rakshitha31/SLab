```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
df = pd.read_csv('BlackFriday.csv')
```

```python
df
```

```python
df.drop(['User_ID', 'Product_ID', 'Stay_In_Current_City_Years'], axis=1)
```

```python
df.City_Category = df.City_Category.fillna('Default')
```

```python
df.City_Category = df.City_Category.map({
    "A": "Metro Cities",
    "B": "Small Towns",
    "C": "Villages"
})
```

```python
df
```

```python
df.Marital_Status = df.Marital_Status.map({
    0: "UnMarried",
    1: "Married"
})
```

```python
df
```

```python
df.rename(columns={"Product_Category_1":"Baseball Caps","Product_Category_2":"Wine tumblers","Product_Category_3":"Pet Raincoats"},inplace=True)
```

```python
ax = sns.countplot(x="Baseball Caps", hue="Gender", palette="Set1", data=df)
ax.set(title="Product Category 1", xlabel="Categories", ylabel="Total")
plt.show()
```

```python
ax = sns.countplot(x="Wine tumblers", hue="Gender", palette="Set1", data=df)
ax.set(title="Product Category 2", xlabel="Categories", ylabel="Total")
plt.show()
```

```python
ax = sns.countplot(x="City_Category", hue="Gender", palette="Set1", data=df)
ax.set(title="City Category", xlabel="Categories", ylabel="Total")
plt.show()
```
